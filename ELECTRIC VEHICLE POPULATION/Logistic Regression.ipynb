**Logistic Regression**
There is a new car in a certain city in a county for a particular year, we need to predict if that car will have electric range greater than 220 or not.

```python
X = df[['County', 'Model Year', 'City', 'Make']]
y = (df['Electric Range']>210).astype(int) # all less than 220 will be 0, else 1
```

```python
X.head(10)
```

```python
X.info()
```

```python
y.head(10)
```

```python
# Let us convert the categorical data to dummy variables
X = pd.get_dummies(X)
```

```python
X.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state =25)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000) # STOP: TOTAL NO. of ITERATIONS REACHED LIMIT. for this error to be rectified
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score
accuracy = accuracy_score(y_test, y_pred)
```

```python
accuracy
```

```python
# Now let us predict the value for a new KIA car that was bought in 2020 in Lynnwood, Snohomish.
new_car = {
    'Model Year': [2020],
    'City': ['Lynnwood'],
    'County': ['Snohomish'],
    'Make': ['KIA']
}
```

```python
new_car_df = pd.DataFrame(new_car)
new_car_df = pd.get_dummies(new_car_df)
new_car_df
```

```python
mew_car_y_pred = model.predict(new_car_df)
```

```python
# So there are these missing columns, Get the column names of the training data
training_columns = set(X_train.columns)
training_columns
```

```python
# Check if there are any columns missing in the new data
missing_columns = training_columns - set(new_car_df.columns)
missing_columns
```

```python
# For all those missing columns, we need to set values as 0
for column in missing_columns:
    new_car_df[column] = 0
```

```python
# Reorder columns to match the training data
new_car_df = new_car_df[X_train.columns]
new_car_df
```

```python
new_car_y_pred = model.predict(new_car_df)
```

```python
new_car_y_pred
```

So we can see that the new car does not have an electric range of more than 220
