```python
holiday = df2[df2['IsHoliday']==True]
No_holiday = df2[df2['IsHoliday']==False]
```

```python
fig,axis = plt.subplots(nrows=2, ncols=1, figsize=(20,7))
sb.lineplot(data=holiday, x='Date' , y='Weekly_Sales', ax=axis[0])
sb.lineplot(data=No_holiday, x='Date' , y='Weekly_Sales', ax=axis[1])
```

```python
plt.figure(figsize=(10,10))
sb.countplot(data=df2, x='Size')
plt.tick_params(axis='x', labelrotation=90)
```

```python
size=df2.groupby(["Store","Size"])["Weekly_Sales"].sum().reset_index()
size['sales_in_thousands'] = size['Weekly_Sales'] / 1000
size['sales_per_store'] = (size['Weekly_Sales']/size['Size'])
size = size.sort_values('sales_per_store', ascending=False).reset_index()
size
```

```python
plt.figure(figsize=(20,20))
sb.lineplot(data=df2, x='Temperature', y='Weekly_Sales')
```

```python
plt.figure(figsize=(10,5))
sb.lineplot(data=size,x="Store",y="sales_in_thousands")
sb.lineplot(data=size,x="Store",y="Size")
plt.tick_params(axis='x', labelrotation=90)
plt.tick_params(axis='y', labelrotation=45)
plt.xticks(range(0,46));
```

```python
plt.figure(figsize=(20,7))
sb.barplot(data=df1,x="Store",y="Weekly_Sales",hue="Type")
plt.tick_params(axis='x', labelrotation=90);
```

```python
figure, axis = plt.subplots(nrows=2, ncols =2, figsize=(15,15))
sb.barplot(y=df2["Weekly_Sales"], x=sorted(df2['Date'].dt.strftime('%Y')),ax=axis[0,0])
sb.barplot(y=df2["Weekly_Sales"], x=sorted(df2['Date'].dt.strftime('%b %Y')),ax=axis[0,1])
sb.barplot(data=df2, y="Weekly_Sales",x=df2["Date"].dt.month,ax=axis[1,0])
sb.barplot(data=df2, y="Weekly_Sales",x=df2["Date"].dt.day,ax=axis[1,1])

axis[0,1].tick_params(axis='x', labelrotation=90)
axis[1,1].tick_params(axis='x', labelrotation=90)
fig.subplots_adjust(bottom=0.07, hspace=0.8)
```
