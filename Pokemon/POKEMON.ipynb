```python

data_groups=data.groupby('type_1')[['hp','attack', 'speed', 'defense', 'special-attack', 'special-defense']].sum()
data_groups['overall_average'] = (data_groups['hp'] + data_groups['attack'] + data_groups['speed'] + data_groups['defense']+ data_groups['special-attack']+ data_groups['special-defense'])/6
```

```python
data2=data.groupby('type_1')
data2['id'].count()
```

```python
data_groups['overall_average_per_pokemon'] = (data_groups['overall_average'])/data2['id'].count()
data_groups = data_groups.reset_index()
```

```python
data_groups
```

```python
plt.figure(figsize=(10,10))
sb.barplot(data=data_groups, y='overall_average_per_pokemon', x='type_1')
plt.tick_params(axis='x', rotation=90)
```
